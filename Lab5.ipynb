```python
import numpy as np
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
import csv
```

```python
data = []
with open("./Market_Basket_Optimisation.csv") as file:
    reader = csv.reader(file, delimiter=',')
    for row in reader:
        data +=[row]
```

```python
data[1:10]
```

```python
len(data)
```

```python
TE = TransactionEncoder()
x = TE.fit_transform(data)
```

```python
x
```

```python
TE.columns_
```

```python
df = pd.DataFrame(x,columns = TE.columns_)
```

```python
df
```

```python
freq_of_itemset = apriori(df, min_support=0.003, use_colnames=True)
```

```python
freq_of_itemset
```

```python
rules = association_rules(freq_of_itemset,metric='lift',min_threshold=0.1)
```
