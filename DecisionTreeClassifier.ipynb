```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
data=pd.read_csv(r"C:\Users\Sanket\Downloads\archive\Admission_Predict.csv")
```

```python
data
```

```python
for i in range(0,400):
    if(data["Chance of Admit "][i]>=0.8):data["Chance of Admit "][i]=1
    else:
        data["Chance of Admit "][i]=0
```

```python
data
```

```python
X=data.iloc[:,1:-2]
Y=data.iloc[:,-1:]
```

```python
Y
```

```python
X
```

```python
Xtrain,Xtest,Ytrain,Ytest=train_test_split(X,Y,train_size=0.75)
```

```python
from sklearn.tree import DecisionTreeClassifier
D_tree=DecisionTreeClassifier(max_depth=3)
D_tree.fit(Xtrain,Ytrain)
```

```python
Y_predict=D_tree.predict(Xtest)
```

```python
Y_predict
```

```python
from sklearn.tree import plot_tree
```

```python
plot_tree(D_tree,fontsize=7,feature_names=list(X.columns),class_names=None,rounded=True,filled=True,max_depth=3)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(Ytest,Y_predict))
```

```python
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay 
```

```python
cm=confusion_matrix(Ytest,Y_predict)
```

```python
print("Confusion Matrix \n",cm)
```

```python
cm_display=ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=[False,True])
cm_display.plot()
```
